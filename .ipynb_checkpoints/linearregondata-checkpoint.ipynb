```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB


```

```python
ds=pd.read_csv("./sales_data_training.csv")
print(ds.shape)
ds["unit_price"]=ds['unit_price'].apply(lambda n: round(n,0))
ds["critic_rating"]=ds['critic_rating'].apply(lambda n:np.ceil(n))
mean=np.mean(ds["total_earnings"])
print(mean)
##ds["total_earnings"]=ds['total_earnings'].apply(lambda n:(n-mean))

ds.head()
```

```python
dk=ds.values
```

```python
y_val=dk[:,8]
print(y_val)
```

```python
df=ds.drop(columns=["total_earnings"])
df.head()
```

```python
x_val=df.values
x_val=np.ceil(x_val)
```

```python
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x_val,y_val)

# Make predictions using the testing set
dj=pd.read_csv("./sales_data_test.csv")
dj["unit_price"]=dj['unit_price'].apply(lambda n: round(n,0))
dj["critic_rating"]=dj['critic_rating'].apply(lambda n:np.ceil(n))
da=dj.values
y_test=da[:,8]
dc=dj.drop(columns=["total_earnings"])
x_test=dc.values

y_pred = regr.predict(x_test)
va=regr.score(x_test,y_test)
print(va)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_val, y_val)
print (dt.score(x_test, y_test))
```

```python
rf = RandomForestRegressor(n_estimators=11)
rf.fit(x_val,y_val)
print (rf.score(x_test, y_test))
```

```python
neigh = KNeighborsRegressor()
neigh.fit(x_val,y_val)
neigh.score(x_test, y_test)
```

```python
preds=[]
clf = GaussianNB()
clf.fit(x_val,y_val)
for ix in range(x_test.shape[0]):
    x_in=x_val[ix,:].reshape(1,-1)
    pred=clf.predict(x_in)
    preds.append(pred)

accu=np.sum(np.array(y_test==preds))/x_test.shape[0]
print(accu)

```
